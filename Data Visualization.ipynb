```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
fruits = ['Apple', 'Banana', 'Coconut', 'Mango', 'Papaya']
sales = [400, 500, 350, 600, 555]
revenue = [350, 400, 325, 400, 425]
```

## Line Plot

```python
plt.figure(figsize=(6, 4)) # 600 X 400 pixels
# [*, s, p, o, x]
# line style (ls) -> dashed, --, solid (default)
plt.plot(fruits, sales, color='red', marker='o', ls='--', label='Sales')
plt.plot(fruits, revenue, color='green', marker='*', label='Revenue')
plt.xlabel('Fruits')
plt.ylabel('Price of Fruits')
plt.title('Total Sales and Revenue of Fruits')
plt.legend(fontsize=8)

# Sales
for idx, val in enumerate(sales):
    plt.text(fruits[idx], val, str(val), ha='center', color='#38594E')

# Revenue
for idx, val in enumerate(revenue):
    plt.text(fruits[idx], val, str(val), ha='center', color='#38594E')

# png, jpeg, jpg
plt.savefig('sales_revenue.png')
plt.show()
```

# Bar Plot

```python
plt.figure(figsize=(6, 4))
plt.bar(fruits, sales, color=['#acd32c', '#e3e3e3', '#2596be', '#6BF2D9', 'orange'],
       width=0.7, edgecolor='black')
plt.xlabel('Fruits')
plt.ylabel('Price of Fruits')
plt.title('Total Sales of Fruits')
# Insert data labels
for idx, val in enumerate(sales):
    plt.text(fruits[idx], val, str(val), ha='center', va='bottom', color='red')
# plt.yticks([])
plt.show()
```

```python
hr = pd.read_csv('Dataset/HR_Dataset Refresh.csv')
hr
```

```python
hr.info()
```

```python
numeric_cols = hr.select_dtypes(include=['float64', 'int64'])
```

```python
numeric_cols = numeric_cols.drop(columns='Zip')
numeric_cols = numeric_cols.drop(columns='EmpID')
```

# Histogram

```python
mean_salary = hr['Salary'].mean()
median_salary = hr['Salary'].median()
```

# Sub-plot

```python
# Find total salary, total absence rate, average employee satisfaction, total employees by departments.
dep_details = hr.groupby('Department').agg(
    total_salary = ('Salary', 'sum'),
    total_absence = ('Absences', 'sum'),
    avg_emp_sat = ('EmpSatisfaction', 'mean'),
    total_employees = ('EmpID', 'count')
).reset_index()
dep_details
```

```python
fig, axes = plt.subplots(2, 2, figsize=(14, 10))
plt.subplots_adjust(hspace=0.5)

# Total Salary
ax1 = axes[0, 0]
ax1.bar(dep_details['Department'], dep_details['total_salary'], color='red', edgecolor='black')
ax1.set_xlabel('Departments')
ax1.set_ylabel('Total Salary')
ax1.set_title('Total Salary by Departments')
ax1.tick_params(axis='x',rotation=15)

for idx, val in enumerate(dep_details['total_salary']):
    ax1.text(dep_details['Department'][idx], val, str(val), ha='center', va='bottom')

# Total Absence Rate
ax2 = axes[0, 1]
ax2.bar(dep_details['Department'], dep_details['total_absence'], color='blue', edgecolor='black')
ax2.set_xlabel('Departments')
ax2.set_ylabel('Total Absences')
ax2.set_title('Total Absence Rate by Departments')
ax2.tick_params(axis='x', rotation=15)

for idx, val in enumerate(dep_details['total_absence']):
    ax2.text(dep_details['Department'][idx], val, str(val), ha='center', va='bottom')

# Average Employee Satisfaction
ax3 = axes[1, 0]
ax3.bar(dep_details['Department'], dep_details['avg_emp_sat'], color='green', edgecolor='black')
ax3.set_xlabel('Departments')
ax3.set_ylabel('Average Employee Satisfaction')
ax3.set_title('Average Employee Satisfaction by Departments')
ax3.tick_params(axis='x', rotation=15)

for idx, val in enumerate(dep_details['avg_emp_sat']):
    fmt_val = round(val, 2)
    ax3.text(dep_details['Department'][idx], fmt_val, str(fmt_val), ha='center', va='bottom')

# Total Employees
ax4 = axes[1, 1]
ax4.bar(dep_details['Department'], dep_details['total_employees'], color='purple', edgecolor='black')
ax4.set_xlabel('Departments')
ax4.set_ylabel('Total Employees')
ax4.set_title('Total Employees by Departments')
ax4.tick_params(axis='x', rotation=15)

for idx, val in enumerate(dep_details['total_employees']):
    ax4.text(dep_details['Department'][idx], val, str(val), ha='center', va='bottom')

plt.savefig('subplot.png')
```
