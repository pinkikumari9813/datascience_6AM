# loc and iloc

```python
'''
Loc -> Location -> (Used for Alphabetical Indexing)
iloc -> Integer Location -> (Used for Numerical Indexing)
'''
```

```python
# dataframe_name[row, column]
personal_details.iloc[0:3, 0:3]
```

```python
personal_details.loc['A':'C', 'Name':'Address']
```
