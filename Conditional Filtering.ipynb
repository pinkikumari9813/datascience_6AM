```python
import pandas as pd
```

```python
df = pd.read_csv('Dataset/Employees.csv')
df
```

```python
'''
Filter data on basis of certain required conditions.
'''
```

```python
# Find details of employees whose salary is in range of 5000 to 8000.
# Find details of employees whose salary is more than 5000 and less than 8000.
emp_sal = (df['SALARY'] >= 5000) & (df['SALARY'] <= 8000)
# df[emp_sal]
df.loc[emp_sal, ['FIRST_NAME', 'LAST_NAME', 'SALARY']]
```

```python
# Between
```

```python
sal_between = df['SALARY'].between(5000, 8000)
df[sal_between]
```

```python
'''
Find details of employees whose salary is in range of 5000 to 8000 
and who must work in HR_REP Job.
'''
emp_sal_job = ((df['SALARY'] >= 5000) & (df['SALARY'] <= 8000)) & (df['JOB_ID'] == 'HR_REP')
df[emp_sal_job]
```

```python
sal = emp_sal = (df['SALARY'] >= 5000) & (df['SALARY'] <= 8000)
df_sal = df[sal]
sal_job = df_sal['JOB_ID'] == 'HR_REP'
```

```python
df_sal[sal_job]
```

```python
# Find employees who works under department 60 and 90.
emp_dep = (df['DEPARTMENT_ID'] == 60) | (df['DEPARTMENT_ID'] == 90)
df[emp_dep]
```

```python
'''
Find employees who works under department 60 and 90 and their salary
must be in range of 6000 to 10000.
'''
emp_in = (df['DEPARTMENT_ID'].isin([60, 90])) & (df['SALARY'].between(6000, 10000))
df[emp_in]
```
