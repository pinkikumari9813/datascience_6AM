```python
import pandas as pd
```

```python
df = pd.read_csv('cleaned_hr_data.csv')
df.head()
```

```python
df.info()
```

```python
# groupby Function
# Aggregate Function
'''
- If we peform analysis using aggregate function then group by is not necessary.
- If we perform analysis using group by then aggregate function must be used.

Aggregate Functions
----------------------
Mean(), Max(), Min(), Count(), Sum()

KPIs (Key Peformance Indicator)
'''
```

```python
# Find total employees by each gender.
gender_count = df.groupby('Gender')['EmpID'].count().reset_index().rename(
    columns = {'EmpID': 'Total Employees'}
)
gender_count
```

```python
# Find total salary distributed to employees by each departments. sum()
dep_sal = df.groupby('Department')['Salary'].sum().reset_index()
dep_sal
```

```python
dep_sal = df.groupby('Department')['Salary'].sum()
# stralign -> String Align -> left, right, center
# numalign -> Number Align -> left, right, center

print(dep_sal.to_markdown(stralign='center', numalign='left'))
```

```python
# Print report of an analysis where total salary, total employees, average employee satisfaction,
# total absence days of emplyoees by each department.
# agg() -> aggregate function

dep_details = df.groupby('Department').agg(
    Total_Salary = ('Salary', 'sum'),
    Total_Employees = ('EmpID', 'count'),
    Avg_Emp_Satisfaction = ('EmpSatisfaction', 'mean'),
    Total_Absence = ('Absences', 'sum')
).sort_values(ascending=True, by='Total_Salary').reset_index()

dep_details['Total_Salary'] = dep_details['Total_Salary'].map("Rs.{:,.2f}".format)
dep_details['Avg_Emp_Satisfaction'] = dep_details['Avg_Emp_Satisfaction'].map("{:,.2f}".format)
```

```python
# abs_data = dep_details['Total_Absence'] < 100
# dep_details[abs_data]
dep_details
```

```python
# Find total employees data by each department and its performance score.
dep_perf = df.groupby(['Department', 'PerformanceScore'])['EmpID'].count().reset_index()
dep_perf
```

```python
pef_detail = (dep_perf['PerformanceScore'] == 'Fully Meets') | (dep_perf['PerformanceScore'] == 'Exceeds')
dep_perf[pef_detail]
```

```python
weekdays = ['Sunday', 'Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday']
# Categorical -> pandas
df['Hire_Day_Name'] = pd.Categorical(df['Hire_Day_Name'], categories=weekdays)
```

```python
# Find total employees hired in each day.
hire_day = df.groupby('Hire_Day_Name', observed=False)['EmpID'].count().reset_index()
hire_day
```
