# Fill Null Data

```python
'''
Mean, Median, Mode, Custom Record
--------------------------------------
Mean and Median -> Numeric Record
Mode -> Categorical
Custom Record for both.
'''
```

```python
# Commission -> Mean, Median
# ManagerId, DepartmentId -> Mode
```

```python
sal_mean = df['COMMISSION_PCT'].mean()
sal_mean
```

```python
# Integer / Float
round_sal_mean = round(sal_mean, 2)
round_sal_mean
```

```python
# Format to string
fmt_sal_mean = f'{sal_mean:,.2f}'
fmt_sal_mean
```

```python
# sal_median = df['COMMISSION_PCT'].median()
# sal_median
```

```python
# Fillna(inplace = True)
# df['COMMISSION_PCT'].fillna(round_sal_mean, inplace=True)
df['COMMISSION_PCT'] = df['COMMISSION_PCT'].fillna(round_sal_mean)
```

# Remove unwanted data from dataset

```python
# use replace() function for removing unwanted data and replacing it from some valid text or spaces.
```

```python
# regex -> (,./!@#$%^&*()_+-=)
# string.replace(find, replace_value, regex=False)
```

```python
df['Employee_Name'] = df['Employee_Name'].str.replace(',', ' ', regex=False)
```

# Remove unwanted space (white-spaces)

```python
# Categorical Records
df['Position'].unique()
```

```python
df['State'].unique()
```

```python
columns = ['Position', 'State', 'Gender', 'MaritalDesc', 'CitizenDesc', 'HispanicLatino', 'RaceDesc',
          'TermReason', 'EmploymentStatus', 'Department', 'RecruitmentSource', 'PerformanceScore']
for col in columns:
    u_col = df[col].unique()
    print(col)
    print('-'*55)
    print(f'{u_col}\n')
```

```python
# Strip()
strip_columns = ['Position', 'Gender', 'Department']
for st_col in strip_columns:
    df[st_col] = df[st_col].str.strip()
```

# Label Encoding

```python
# Hispanic Latino
hl_data = {
    'No': 'No',
    'Yes': 'Yes',
    'no': 'No',
    'yes': 'Yes'
}
df['HispanicLatino'] = df['HispanicLatino'].map(hl_data)
```
